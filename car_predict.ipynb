## What to clean

- year is not in int and also has non-year values
- Not in float or integer also non price values
- kms has kms written in them also comma, not in integer
- fuel_type has nan values
- keep first 3 words of name

## Cleaning of Data

```python
backup=car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

```python
car=car[car["Price"]!="Ask For Price"]
```

```python
# replace comma with empty sring
```

```python
car['Price']=car["Price"].str.replace(',','').astype(int)
```

```python
car.info()
```

```python
car.info()
```

```python
car['kms_driven'].str.split(' ').str.get(0).str.replace(',',' ')
```

```python
car.info()
```

```python
car.info()
```

```python
# Remove commas, 'kms' text, and convert to integer
car['kms_driven'] = car['kms_driven'].str.replace(',', '')  # Remove commas
car['kms_driven'] = car['kms_driven'].str.replace('kms', '')  # Remove 'kms'
car['kms_driven'] = car['kms_driven'].str.strip()  # Remove extra spaces
car['kms_driven'] = pd.to_numeric(car['kms_driven'], errors='coerce')  # Convert to number
car = car[~car['kms_driven'].isna()]  # Remove rows where conversion failed
car['kms_driven'] = car['kms_driven'].astype(int)  # Convert to integer
```

```python
car.info()
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car.reset_index(drop=True)
```

```python
car.describe()
```

## OUTLIERS

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

```python
X=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.30,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.30,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
